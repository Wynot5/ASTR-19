```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = np.random.binomial(10, 0.5, 1000)

plt.hist(data, bins=100, alpha=0.5, edgecolor="black")
plt.xlabel("x")
plt.ylabel("Per bin")

plt.savefig("Journal2-4.pdf", bbox_inches="tight", dpi=600)

plt.show
```
