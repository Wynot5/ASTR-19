```python
import numpy as np
from tabulate import tabulate
```

```python
def sine(x):
    return np.sin(x)
```

Returns the sin of x

```python
def cosine(x):
    return np.cos(x)
```

Returns the cos of x

```python
twopi = 2*np.pi

x = np.linspace(0, twopi, 1000)
s = sine(x)
c = cosine(x)

data = [(a, b, c) for a, b, c in zip(x, s, c)]
headers = ["x", "sin(x)", "cos(x)"]

table = tabulate(data, tablefmt="grid", headers=headers, floatfmt=".3f")
```

```python
new_data = [];
for i in range(10):
    new_data.append(data[i])

print(tabulate(new_data, tablefmt="grid", headers=headers, floatfmt=".3f"))
```
